```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
%matplotlib inline
```

# Regresiion Plots

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex')
```

# Working with Markers

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex' ,palette='coolwarm',markers='*',scatter_kws={'s':10})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex' ,markers=['*','^'],scatter_kws={'s':50})
```

# Using a Grid

```python
# this is creating 2 coulmns of sex=> male and female
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,row='sex',col='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex')
```

# Aspect and Size

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',aspect=0.7,size=5)
```
