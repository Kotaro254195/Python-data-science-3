# Exercise 3

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,.4,.4])

x=np.arange(0,110,step=10) 
y=np.arange(0,220,step=20)
z=y**2

axes1.plot(x,z,'b')
axes2.plot(x,y,'b')

plt.xlim(0,100)
plt.ylim(0,200)
```
